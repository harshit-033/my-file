```python
###modeling a bank account
class bank_account:
  def __init__(self,owner,balance):
    self.owner=owner
    self.balance=balance

  def deposite(self,amount):
    self.balance=self.balance+amount
    print(f"balance is {self.balance}")

  def withdraw(self,amount):
    if self.balance>=amount:
      self.balance=self.balance-amount
      print(f"balance is {self.balance}")
    else:
      print("insufficient balance")
  def balance_check(self):
    print(f"balance is {self.balance}")


account1=bank_account("p1",1000)
account1.deposite(200)
account1.withdraw(1500)
account1.balance_check()
```

```python
class car:
  def __init__(self,windows,doors,enginetype):
    self.windows=windows
    self.doors=doors
    self.enginetype=enginetype

  def drive(self):
    print(f"{self.enginetype} car is driving")

  def stop(self):
    print("car is stopped")
```

```python
car1=car(4,2,"petrol")
car1.drive()
car1.stop()
```

```python
class audi(car):
  def __init__(self,windows,doors,enginetype,ai):
    super().__init__(windows,doors,enginetype)
    self.ai=ai

  def is_ai_car(self):
    print(f"audi is self driving ai car: {self.ai}")


```

```python
audi_1=audi(4,2,"electric",True)
audi_1.is_ai_car()
```

```python
audi_1.drive()
```

```python
class animal:
  def __init__(self,name,age):
    self.name=name
    self.age=age
```

```python
class pet:
  def __init__(self,owner):
    self.owner=owner

  def show_owner(self):
    print(f"owner is {self.owner}")
```

```python
class dog(animal,pet):
  def __init__(self,name,age,owner,healthy):
    animal.__init__(self,name,age)
    pet.__init__(self,owner)
    self.healthy=healthy

  def show_healthy(self):
    print(f"dog is healthy: {self.healthy}")
```

```python
dog1=dog("tom",2,"p1",True)
dog1.show_owner()
dog1.show_healthy()
```

```python
class shape:
  def area(self):
    return "the area of the figure"

class rectangle(shape):
  def __init__(self,length,width):
    self.length=length
    self.width=width

  def area(self):
    return self.length*self.width

class circle(shape):
  def __init__(self,radius):
    self.radius=radius

  def area(self):
    return 3.14*self.radius**2

def get_area(obj):
  return obj.area()

rectangle1=rectangle(4,5)
circle1=circle(4)
print(get_area(rectangle1))
print(get_area(circle1))
```

```python
from abc import ABC,abstractmethod

class Vehicle(ABC):
  @abstractmethod
  def start_engine(self):
    pass

class Car(Vehicle):
  def start_engine(self):
    print("car engine started")

class Motorcycle(Vehicle):
  def start_engine(self):
    print("motorcycle engine started")

#objects of car and motorcycle
car=Car()
Motorcycle=Motorcycle()

#calling the start_engine method
car.start_engine()
Motorcycle.start_engine()
```
